```python
import requests

url = "https://www.ncei.noaa.gov/access/billions/state-summary/US"

response = requests.get(url)
response
from bs4 import BeautifulSoup


soup_doc = BeautifulSoup(response.content, 'html.parser')
soup_doc.prettify()
```

```python
tables = soup_doc.find_all("table")
```

```python
for col, table in enumerate(tables):
    table_data = []

    for row in table.find_all("tr"):
        cells = row.find_all(["th", "td"])
        row_data = [cell.get_text(strip=True) for cell in cells]
        if row_data:
            table_data.append(row_data)

    filename = f"ncei_us_table_{col+1}.csv"
    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerows(table_data)

    print(f"{filename}")
```

```python
import pandas as pd

df = pd.read_csv("ncei_us_table_1.csv")

def extract_number(val):
    import re
    try:
        match = re.search(r"-?\d+\.?\d*", str(val).replace(",", ""))
        return float(match.group()) if match else None
    except:
        return None

for col in df.columns[1:]:
    df[col] = df[col].apply(extract_number)

df.to_csv("ncei_us_table_1_cleaned.csv", index=False)
print("Done.")
```
