```python
import requests

url = "https://www.evergladesuniversity.edu/blog/worst-disasters-in-us-history/"

response = requests.get(url)
response
from bs4 import BeautifulSoup


soup_doc = BeautifulSoup(response.content, 'html.parser')
soup_doc.prettify()
```

```python
par = soup_doc.find_all("h2")
par
```

```python
data = []
for h2 in soup_doc.find_all("h2"):
    disaster_name = h2.get_text(strip=True)
    description_parts = []
    for sibling in h2.find_next_siblings():
        if sibling.name == "h2":
            break
        if sibling.name == "p":
            description_parts.append(sibling.get_text(strip=True))
    
    description = " ".join(description_parts)
    data.append({"name of disaster": disaster_name, "description": description})
with open("disasters.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=["name of disaster", "description"])
    writer.writeheader()
    writer.writerows(data)
```
